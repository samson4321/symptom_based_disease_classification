```python
df = pd.get_dummies(df, columns=['Gender'])
```

```python
bool_cols = df.select_dtypes(include="bool").columns
df[bool_cols] = df[bool_cols].astype(int)
```

```python
# 1. Transformer la colonne 'Symptoms' en colonnes one-hot
symptoms = df['Symptoms'].str.get_dummies(sep=', ')

# 2. Supprimer la colonne texte 'Symptoms' du DataFrame original
df = df.drop('Symptoms', axis=1)

# 3. Concatner les colonnes symptmes one-hot au DataFrame
df = pd.concat([df, symptoms], axis=1)

# 4. Dfinir X et y
X = df.drop('Disease', axis=1)
y = df['Disease']

# 5. Split train/test
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

# 6. Scaling (toutes les colonnes sont maintenant numriques)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Puis tu peux continuer avec l'entranement et l'valuation des modles comme avant
```

```python
df.head()
```

```python
def test_knn(k_values):
    print("----- Test KNN -----")
    for k in k_values:
        knn = KNeighborsClassifier(n_neighbors=k)
        knn.fit(X_train_scaled, y_train)
        y_pred = knn.predict(X_test_scaled)
        acc = accuracy_score(y_test, y_pred)
        print(f"K={k} --> Accuracy: {acc:.4f}")

def test_random_forest():
    print("\n----- Test Random Forest -----")
    rf = RandomForestClassifier(random_state=42)
    rf.fit(X_train_scaled, y_train)
    y_pred = rf.predict(X_test_scaled)
    acc = accuracy_score(y_test, y_pred)
    print(f"Random Forest Accuracy: {acc:.4f}")
    print("\nClassification Report:\n", classification_report(y_test, y_pred))


# Tester diffrents k pour KNN
k_values = [3, 5, 7, 9, 11]
test_knn(k_values)

# Tester Random Forest
test_random_forest()
```

```python
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

train_acc = []
test_acc = []
k_values = range(1, 11)

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_scaled, y_train)

    train_acc.append(knn.score(X_train_scaled, y_train))
    test_acc.append(knn.score(X_test_scaled, y_test))

# Plot
plt.figure(figsize=(8, 5))
plt.plot(k_values, train_acc, label="Training accuracy", marker="o")
plt.plot(k_values, test_acc, label="Test accuracy", marker="o", linestyle="--")

plt.xlabel("Number of neighbors (k)")
plt.ylabel("Accuracy")
plt.title("Training vs Test Accuracy for KNN")
plt.legend()
plt.grid(True)
plt.show()
```
